# *Basic Machine Learning Code*

```python
import sklearn 
from sklearn.linear_model import LinearRegression
model = LinearRegression()
x = [[1],[2],[3],[4]]
y = [2,4,6,8]
model.fit(x,y)
prediction = model.predict([[6]])
print("Prediction for input 6 is : ",prediction)
```

# *Visualizing the above code*

```python
import matplotlib.pyplot as plt 
plt.scatter(x,y, color = 'red', label = 'Data Points')
plt.plot(x, model.predict(x), color = 'green', label = "Regression Line")
plt.legend(loc = 'center left', bbox_to_anchor = (1,0.5))
plt.title("Linear Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.tight_layout()
plt.show()
```

# *Predicting the value of (y) for a user input (x)*

```python
user_value = float(input("Enter a number to predict : "))
prediction = model.predict([[user_value]])
print("Prediction is : ", prediction)
```

## To predict the price of the banana based on the weights given . 

```python
weights = np.array([1,2,3,4]).reshape(-1,1)
prices = np.array([30,60,90,120])
plt.scatter(weights,prices, color = 'red',label = 'Weights')
plt.plot(weights, prices, color = 'green', label = "Price")
plt.legend(loc = 'center left', bbox_to_anchor = (1,0.5))
plt.title("Plot")
plt.show()
```

```python
model.fit(weights, prices)
user_value = float(input("Enter a number to predict : "))
prediction = model.predict([[user_value]])
print(f"The price of the banana weighing {user_value} is {prediction}")
plt.scatter(weights, prices, color = 'red', label = 'Weights')
plt.plot(weights, model.predict(weights), color = 'green', label = 'Price')
plt.legend(loc = 'center left', bbox_to_anchor = (1,0.5))
plt.title("Linear Regression Line")
plt.xlabel("Weights")
plt.ylabel("Prices")
plt.tight_layout()
plt.show()
```
