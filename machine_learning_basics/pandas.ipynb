Pandas is a Python library used for working with structured data (like tables, CSVs, Excel sheets).

## Working with DataFrames

```python
import pandas as pd 
```

```python
s = pd.Series([10,20,30,40]) # Series represents a one dimensional labeled array 
print(s)
```

```python
s = pd.Series([100,200,300], index = ['A','B','C']) # Renaming the indexes
print(s)
print(s['B'])
```

```python
data = {'Name':['Arya','Ravi','Meera'], # Converting dictionary data into a table
        'Marks':[85,92,78],
        'City':['Pune','Delhi','Mumbai']}
df = pd.DataFrame(data)
df
```

```python
df1 = df.copy() # Creates a copy of the table 'df' into 'df1'
df1['Grades']=(['A','A+','B'])
df1
```

```python
print(df)
print()
print(df1)
```

```python
del df1['Grades']
print(df1)
```

```python
print(df)
print()
print(df1)
```

```python
df.loc[1] # Access the record at the particular index location using 'loc'
```

```python
df.iloc[0:2] # Iterates over the specified range & prints it
```

```python
df[['Name','City']] # Accessing particular columns
```

```python
print(df.loc[1])
```
