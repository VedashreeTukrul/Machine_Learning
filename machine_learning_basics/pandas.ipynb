## Working with DataFrames

```python
import pandas as pd 
```

```python
s = pd.Series([10,20,30,40]) # Series represents a one dimensional labeled array 
print(s)
```

```python
s = pd.Series([100,200,300], index = ['A','B','C']) # Renaming the indexes
print(s)
print(s['B'])
```

```python
data = {'Name':['Arya','Ravi','Meera'], # Converting dictionary data into a table
        'Marks':[85,92,78],
        'City':['Pune','Delhi','Mumbai']}
df = pd.DataFrame(data)
df
```

```python
df1 = df.copy() # Creates a copy of the table 'df' into 'df1'
df1['Grades']=(['A','A+','B'])
df1
```

```python
print(df)
print()
print(df1)
```

```python
del df1['Grades']
print(df1)
```

```python
print(df)
print()
print(df1)
```

## Working with Excel Tables (csv's)

#### Load the dataset

```python
df = pd.read_csv("recharge.csv")
```

#### Basic Operations

```python
print(df.head()) #shows first 5 rows
```

```python
print(df.to_string()) # displays all records
```

#### View Data Information

```python
df.shape # (rows, columns)
```

```python
df.columns # Column names
```

```python
df.info() # Data types and nulls
```

```python
df.describe() #Summary stats
```

#### Select Columns and Rows

```python
df[['Region']] # Single column
```

```python
df[['Region','amount']] # Multiple column
```

```python
df.loc[0] # Row by index
```

```python
df.iloc[0:3] # First 3 rows
```

#### Basic Calculations

```python
df['amount'].sum()
```

```python
df['amount'].mean()
```

```python
df['amount'].max()
```

```python
df['plan_name'].value_counts()
```

#### Group and Aggregate 

```python
df.groupby('Region')['amount'].sum()
```

```python
df.groupby(['Region', 'plan_name'])['amount'].mean()
```

### *Data Cleaning and Analysis*

### Handle Missing Values

```python
df.isnull().sum() # Counts missing in each column
```

```python
df.dropna() # Drops rows with missing values
```

```python
df.fillna(0) # Replace the missing values with 0
```

#### Rename Columns

```python
df.rename(columns={'amount':'Recharge_Amount'}, inplace = True)
df
```

#### Filter Rows

```python
df[df['Recharge_Amount'] > 300] # High-value recharges
```

#### Create New Columns

```python
df['Bonus'] = df['Recharge_Amount'] * 0.10
df
```

```python
df['Recharge_Flag'] = df['Recharge_Amount'] > 299
df
```

#### Sort Data

```python
df.sort_values('Recharge_Amount',ascending = False)
```

#### Export Cleaned Data

```python
df.to_csv("cleaned_recharge_data.csv", index = False)
```
